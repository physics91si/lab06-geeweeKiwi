```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from matplotlib.backends.backend_pdf import PdfPages
```

# Part 1

```python
# TODO define gaussian function here
def wavepacket(x, k, sigma, mu = 0, h=1):
    """
    This function creates a wavepacket on the interval defined by x with
    wavevector k and standard deviation sigma.
    """
    return h * np.sin(k*x) *  np.exp(-((x-mu)**2)/(2*sigma**2))
xvals = np.arange(-5,5, 0.25)
pckt1 = wavepacket(xvals, 5, 3)
pckt2 = wavepacket(xvals, 4, .5)
pckt3 = wavepacket(xvals, 20, 5)
pckt4 = wavepacket(xvals, 2, 3)
```

```python
# TODO plot gaussians here
plt.plot(xvals, pckt1, 'b:', xvals, pckt2, 'r--', xvals, pckt3, 'g-.', xvals, pckt4, 'm')
plt.show()
```

```python
# TODO fit gaussians, with known parameters, here
params = curve_fit(wavepacket, xvals, pckt1)

# TODO plot result
plt.plot(xvals, wavepacket(xvals, *params[0]))
plt.plot(xvals, pckt1, 'r:')
plt.show()
```

```python
# TODO: load data here
data = np.loadtxt('HDalpha_data')
data = np.array(data)
x_data = data[:,0]
y_data = data[:,1]
```

```python
# TODO: plot it
plt.plot(x_data, y_data)
plt.show()
```

```python
# TODO: plot data as emission signal
plt.plot(x_data, -1*y_data)
plt.show()
```

```python
# TODO: fit a function to this data and plot it
e_data = -1*y_data

def doublegauss(x, k1, k2, sigma1, sigma2, mu1, mu2, h1, h2):
    return wavepacket(x, k1, sigma1, mu1, h1) + wavepacket(x, k2, sigma2, mu2, h2)

params2 = curve_fit(doublegauss, x_data, e_data, p0=[1, 1, 1, 1, 13130, 13130, .001, .001])

plt.plot(x_data, doublegauss(x_data, *params2[0]), label = 'Double Gaussian Curve')
plt.plot(x_data, e_data , 'r:', label = 'data')
plt.legend(loc = 'upper right')
plt.xlim([13124,13138])
plt.title('Plot of Hydrogen/Deutirium Emissions')
plt.ylabel('Emission Levels')
plt.xlabel('x axis')
plt.show()
#plt.close()
```

```python
def flux_for_dipole(x_pos, y_pos, Q_vals, X, Y):
    """
    Returns the flux based on the information of the charges
    and the grid of the plot.
    Code modified from Physics 113, written by Blas Cabrera
    """
    flux = np.zeros((len(X), len(Y)))
    x_m = np.outer(np.ones((X.size)), X); 
    y_m = np.outer(Y, np.ones((Y.size))); 
    for n in range(len(Q_vals)):
        flux += 2*np.pi*Q_vals[n]*(1 - (x_pos[n]-x_m)/np.sqrt((y_pos[n]-y_m)**2+(x_pos[n]-x_m)**2))
    return flux
```

```python
# TODO create arrays of information
x_pos = np.array([20,5])
y_pos = np.array([-4,15])
charges = np.array([15,20])
X = np.arange(-31.5,32, 1)
Y = X
# TODO get flux
flux = flux_for_dipole(x_pos, y_pos, charges, X, Y)
# TODO plot contours of the flux
plt.contour(X,Y, flux, 10)
plt.scatter(x_pos,y_pos, color = 'r', s = 10*charges)
plt.show()
```
